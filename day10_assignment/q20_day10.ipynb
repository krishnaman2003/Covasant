```python
#1. Read and check data 
path = r"code\data\titanic_train.csv"
titanic = pd.read_csv(path)
titanic.head()
```

```python
print(titanic.head())
```

```python
#2. Which gender survived more
survival_by_gender = titanic.groupby('sex')['survived'].mean()
print(survival_by_gender)
```

```python
titanic.groupby("sex")['survived'].sum().idxmax()
```

```python
#3. Does it depend on passenger class?
survival_pivot = titanic.pivot_table(
    index='pclass',
    columns='sex',
    values='survived',
    aggfunc='mean'
)
survival_percent = survival_pivot * 100
survival_percent
```

```python
#4. can we see % of survival of each gender and pclass
survival_percent.plot(kind='bar', figsize=(8,5))
plt.xlabel('Passenger Class')
plt.ylabel('Survival Percentage (%)')
plt.title('Titanic Survival by Passenger Class and Gender')
plt.legend(title='Gender')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#1. Read and check data
path = r"code\data\federer.csv"
federer = pd.read_csv(path, parse_dates=['start date'])
federer.head()
```

```python
print(federer.head())
```

```python
#2. How many % of matches won by our player? ('winner')
player = 'Roger Federer'
total_matches = len(federer)
wins = (federer['winner'] == player).sum()
win_pct = wins / total_matches * 100
print(f"{player} won {win_pct:.4f}% of matches.")
```

```python
'''3. Proportion of double faults wrt total points in each match 
This number is an indicator of the player's state of mind, his level of self-confidence, 
his willingness to take risks while serving, and other parameters.
columns:
'player1 double faults' and 'player1 total points total'
Display simple stats of above'''

federer['double_fault_rate'] = (
    federer['player1 double faults'] / federer['player1 total points total']
)
federer['double_fault_rate']
```

```python
double_fault_stats = federer['double_fault_rate'].describe()
print(double_fault_stats)
```

```python
avg_win_surface = federer.groupby('surface')['winner'].apply(
    lambda winners: (winners == player).mean() * 100
)
print(avg_win_surface)
```

```python
'''5. Display the proportion of double faults as a function of the tournament date, 'start date' 
Trend: display average double faults in each year''' 

federer['start date'] = pd.to_datetime(federer['start date'], dayfirst=True)
federer['year'] = federer['start date'].dt.year
federer['proportion'] = (federer['player1 double faults'] /
                         federer['player1 total points total'])
avg_double = federer.groupby('year')['proportion'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_double.index, avg_double.values, marker='o')
plt.title('Average Double Fault Proportion by Year')
plt.xlabel('Year')
plt.ylabel('Average Proportion of Double Faults')
plt.grid(True)
plt.tight_layout()
plt.show()
```
